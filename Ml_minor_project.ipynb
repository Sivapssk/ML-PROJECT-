```python




import pandas as pd
import numpy as np
!pip install optuna
import optuna
from sklearn.model_selection import train_test_split, StratifiedKFold
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier, StackingClassifier, VotingClassifier
from xgboost import XGBClassifier
from imblearn.over_sampling import SMOTE
from sklearn.metrics import accuracy_score
from sklearn.compose import ColumnTransformer
import joblib

# Load dataset
df = pd.read_csv('/content/drive/MyDrive/ml project/loan_prediction.csv')



# Handling missing values for Numerical features with median
numerical_features = df.select_dtypes(include=np.number).columns
df[numerical_features] = df[numerical_features].fillna(df[numerical_features].median())

# Handling missing values for Categorical features with mode
categorical_features = df.select_dtypes(include=['object']).columns.drop(['Loan_ID', 'Loan_Status'])
for feature in categorical_features:
    df[feature] = df[feature].fillna(df[feature].mode()[0])

# Encode target variable
df['Loan_Status'] = df['Loan_Status'].map({'Y': 1, 'N': 0})

# Define X (features) and y (target)
X = df.drop(columns=['Loan_Status', 'Loan_ID'])
y = df['Loan_Status']

# Preprocessing using ColumnTransformer
categorical_features = X.select_dtypes(include=['object']).columns.tolist()
numerical_features = X.select_dtypes(exclude=['object']).columns.tolist()
preprocessor = ColumnTransformer([
    ('num', StandardScaler(), numerical_features),
    ('cat', OneHotEncoder(handle_unknown='ignore'), categorical_features)])

# Apply preprocessing and SMOTE
X_processed = preprocessor.fit_transform(X)
X_resampled, y_resampled = SMOTE(random_state=42).fit_resample(X_processed, y)

# Splitting data
X_train, X_test, y_train, y_test = train_test_split(X_resampled, y_resampled, test_size=0.2, random_state=42, stratify=y_resampled)

# Optuna for Hyperparameter Tuning (XGBoost)
def objective(trial):
    params = {
        'n_estimators': trial.suggest_int('n_estimators', 200, 1200),
        'max_depth': trial.suggest_int('max_depth', 4, 12),
        'learning_rate': trial.suggest_loguniform('learning_rate', 0.005, 0.3),
        'subsample': trial.suggest_float('subsample', 0.7, 1.0),
        'colsample_bytree': trial.suggest_float('colsample_bytree', 0.7, 1.0),
        'gamma': trial.suggest_float('gamma', 0, 1),
        'reg_alpha': trial.suggest_float('reg_alpha', 0, 1),
        'reg_lambda': trial.suggest_float('reg_lambda', 0, 1),
    }
    model = XGBClassifier(**params, random_state=42)
    model.fit(X_train, y_train)
    preds = model.predict(X_test)
    return accuracy_score(y_test, preds)

study = optuna.create_study(direction="maximize")
study.optimize(objective, n_trials=75)  # Increased trials for better optimization

best_params = study.best_params
print("Best Hyperparameters:", best_params)

# Train best XGBoost model
best_xgb = XGBClassifier(**best_params, random_state=42)
best_xgb.fit(X_train, y_train)
y_pred_xgb = best_xgb.predict(X_test)
accuracy_xgb = accuracy_score(y_test, y_pred_xgb)
print(f"XGBoost Accuracy: {accuracy_xgb}")

# Save best model
joblib.dump(best_xgb, "/content/best_loan_status_model.pkl")
print("Model saved successfully!")
```

```python
!pip install gradio
!pip install xgboost
!pip install imblearn
!pip install scikit-learn
!pip install pandas

import gradio as gr
import joblib
import pandas as pd
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer

# Load trained model
model = joblib.load('/content/best_loan_status_model.pkl')

# Define categorical and numerical features
categorical_features = ['Gender', 'Married', 'Education', 'Self_Employed', 'Property_Area', 'Credit_History']
numerical_features = ['Dependents', 'ApplicantIncome', 'CoapplicantIncome', 'LoanAmount', 'Loan_Amount_Term']

# Load your training data (used to fit preprocessor)
training_data = pd.read_csv('/content/drive/MyDrive/ml project/loan_prediction.csv')
training_data['Dependents'] = pd.to_numeric(training_data['Dependents'], errors='coerce').fillna(0).astype(int)
X_train = training_data[categorical_features + numerical_features]

# Create preprocessor and fit
preprocessor = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), numerical_features),
        ('cat', OneHotEncoder(sparse_output=False, handle_unknown='ignore'), categorical_features)
    ])
preprocessor.fit(X_train)  # Fit on the entire training data (including Dependents)

# Prediction function
def predict_loan_status(Gender, Married, Dependents, Education, Self_Employed, ApplicantIncome, CoapplicantIncome, LoanAmount, Loan_Amount_Term, Credit_History, Property_Area):
    try:
        # Create input DataFrame
        input_data = pd.DataFrame({
            'Gender': [Gender],
            'Married': [Married],
            'Dependents': [Dependents],
            'Education': [Education],
            'Self_Employed': [Self_Employed],
            'ApplicantIncome': [ApplicantIncome],
            'CoapplicantIncome': [CoapplicantIncome],
            'LoanAmount': [LoanAmount],
            'Loan_Amount_Term': [Loan_Amount_Term],
            'Credit_History': [Credit_History],
            'Property_Area': [Property_Area]

        })

        # Preprocess input data - Convert 'Dependents' to numeric and handle errors
        input_data['Dependents'] = pd.to_numeric(input_data['Dependents'], errors='coerce').fillna(0).astype(int)

        # Apply preprocessing using the fitted preprocessor
        input_data_processed = preprocessor.transform(input_data)

        # Make prediction
        result = model.predict(input_data_processed)[0]
        return "âœ… Loan Approved" if result == 1 else "âŒ Loan Not Approved"

    except Exception as e:
        return f"âš ï¸ Error: {e}"

# Creating the Gradio Interface
iface = gr.Interface(
    fn=predict_loan_status,
    inputs=[
        gr.Radio(choices=["Male", "Female"], label="Gender"),
        gr.Radio(choices=["Yes", "No"], label="Married"),
        gr.Radio(choices=["0", "1", "2", "3"], label="Dependents [0,1,2,3+]"),
        gr.Radio(choices=["Graduate", "Not Graduate"], label="Education"),
        gr.Radio(choices=["Yes", "No"], label="Self_Employed"),
        gr.Number(label="Applicant Income"),
        gr.Number(label="Coapplicant Income"),
        gr.Number(label="Loan Amount"),
        gr.Number(label="Loan Amount Term"),
        gr.Radio(choices=["Yes", "No"], label="Credit History"),
        gr.Radio(choices=["Rural", "Urban", "Semiurban"], label="Property Area")
    ],
    outputs="text",
    title="ðŸ¦ Loan Status Prediction",
    description="Enter the details to check if the loan will be **Approved or Not Approved**."
)

# Launch the interface
iface.launch()
```

```python
!pip install optuna
!pip install catboost
!pip install xgboost
!pip install imblearn

import pandas as pd
import numpy as np
import optuna
from sklearn.model_selection import train_test_split, StratifiedKFold
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier, StackingClassifier, VotingClassifier
from xgboost import XGBClassifier
from catboost import CatBoostClassifier
from imblearn.over_sampling import SMOTE
from sklearn.metrics import accuracy_score
from sklearn.compose import ColumnTransformer
import joblib

# Load dataset
df = pd.read_csv('/content/drive/MyDrive/ml project/loan_prediction.csv')

# Handling missing values for Numerical features with median
numerical_features = df.select_dtypes(include=np.number).columns
df[numerical_features] = df[numerical_features].fillna(df[numerical_features].median())

# Handling missing values for Categorical features with mode
categorical_features = df.select_dtypes(include=['object']).columns.drop(['Loan_ID', 'Loan_Status'])
for feature in categorical_features:
    df[feature] = df[feature].fillna(df[feature].mode()[0])

# Encode target variable
df['Loan_Status'] = df['Loan_Status'].map({'Y': 1, 'N': 0})

# Define X (features) and y (target)
X = df.drop(columns=['Loan_Status', 'Loan_ID'])
y = df['Loan_Status']

# Preprocessing using ColumnTransformer
categorical_features = X.select_dtypes(include=['object']).columns.tolist()
numerical_features = X.select_dtypes(exclude=['object']).columns.tolist()

# Load or recreate your training data to fit the preprocessor:
training_data = df.copy()  # Make a copy of the original data

preprocessor = ColumnTransformer([
    ('num', StandardScaler(), numerical_features),
    ('cat', OneHotEncoder(sparse_output=False, handle_unknown='ignore'), categorical_features)])

# Apply preprocessing and SMOTE
X_processed = preprocessor.fit_transform(training_data[categorical_features + numerical_features])  # Fit on training data
X_resampled, y_resampled = SMOTE(random_state=42).fit_resample(X_processed, y)

# Splitting data
X_train, X_test, y_train, y_test = train_test_split(X_resampled, y_resampled, test_size=0.2, random_state=42, stratify=y_resampled)

# Optuna for Hyperparameter Tuning (XGBoost)
def objective(trial):
    params = {
        'n_estimators': trial.suggest_int('n_estimators', 200, 1200),
        'max_depth': trial.suggest_int('max_depth', 4, 12),
        'learning_rate': trial.suggest_loguniform('learning_rate', 0.005, 0.3),
        'subsample': trial.suggest_float('subsample', 0.7, 1.0),
        'colsample_bytree': trial.suggest_float('colsample_bytree', 0.7, 1.0),
        'gamma': trial.suggest_float('gamma', 0, 1),
        'reg_alpha': trial.suggest_float('reg_alpha', 0, 1),
        'reg_lambda': trial.suggest_float('reg_lambda', 0, 1),
    }
    model = XGBClassifier(**params, random_state=42)
    model.fit(X_train, y_train)
    preds = model.predict(X_test)
    return accuracy_score(y_test, preds)

study = optuna.create_study(direction="maximize")
study.optimize(objective, n_trials=75)  # Increased trials for better optimization

best_params = study.best_params
print("Best Hyperparameters:", best_params)

# Train best XGBoost model
best_xgb = XGBClassifier(**best_params, random_state=42)
best_xgb.fit(X_train, y_train)
y_pred_xgb = best_xgb.predict(X_test)
accuracy_xgb = accuracy_score(y_test, y_pred_xgb)
print(f"XGBoost Accuracy: {accuracy_xgb}")

# Save best model
joblib.dump(best_xgb, "/content/best_loan_status_model.pkl")
print("Model saved successfully!")

# Now, create your test data as a pandas DataFrame:
test_data = pd.DataFrame({
    'Gender': ['Male'],  # Provide string values
    'Married': ['Yes'],  # Provide string values
    'Dependents': ['2'],   # Provide string values
    'Education': ['Graduate'],  # Provide string values
    'Self_Employed': ['No'],  # Provide string values
    'ApplicantIncome': [100000],
    'CoapplicantIncome': [50000],
    'LoanAmount': [5000],
    'Loan_Amount_Term': [180],
    'Credit_History': ['1'],  # Provide string values
    'Property_Area': ['Urban']  # Provide string values
}, index=[0])


# Preprocess your test data:
test_data_processed = preprocessor.transform(test_data)

# Perform prediction:
prediction = model.predict(test_data_processed)

print("Test Prediction:", prediction)
```
